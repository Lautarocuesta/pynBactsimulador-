```python
import numpy as np
import matplotlib.pyplot as plt

#  Entrada de datos
h0 = float(input("Ingrese la altura inicial (m): "))
t_total = np.sqrt(2 * h0 / 9.8)  # Calculamos el tiempo total de cada
t = np.linspace(0, t_total, 100)  # Generamos puntos de tiempo en intervalos pequeos

#  Parmetros fsicos
g = 9.8  # Aceleracin de la gravedad (m/s)

#  Clculo de posicin, velocidad y aceleracin
y = h0 - (1/2) * g * t**2  # Posicin (m)
v = -g * t  # Velocidad (m/s)
a = -g * np.ones_like(t)  # Aceleracin (m/s) (constante)

#  Grficos
fig, axs = plt.subplots(3, 1, figsize=(8, 10), sharex=True)
plt.suptitle("Grficos de Cada Libre", fontsize=16, fontweight='bold')

#  Grfico de posicin vs. tiempo
axs[0].plot(t, y, color='royalblue', linewidth=2)
axs[0].invert_yaxis()  # Invertimos el eje para representar cada
axs[0].set_ylabel("Posicin (m)", fontsize=12)
axs[0].grid(True, linestyle='--', alpha=0.7)
axs[0].set_title("Posicin vs. Tiempo", fontsize=14)

#  Grfico de velocidad vs. tiempo
axs[1].plot(t, v, color='crimson', linewidth=2)
axs[1].set_ylabel("Velocidad (m/s)", fontsize=12)
axs[1].grid(True, linestyle='--', alpha=0.7)
axs[1].set_title("Velocidad vs. Tiempo", fontsize=14)

#  Grfico de aceleracin vs. tiempo
axs[2].plot(t, a, color='forestgreen', linewidth=2)
axs[2].set_xlabel("Tiempo (s)", fontsize=12)
axs[2].set_ylabel("Aceleracin (m/s)", fontsize=12)
axs[2].grid(True, linestyle='--', alpha=0.7)
axs[2].set_title("Aceleracin vs. Tiempo", fontsize=14)

#  Ajustes finales
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  Entrada de datos
v0 = float(input("Ingrese la velocidad inicial (m/s): "))
t_total = (2 * v0) / 9.8  # Tiempo total de subida y bajada
t = np.linspace(0, t_total, 100)  # Generamos puntos de tiempo en intervalos pequeos

#  Parmetros fsicos
g = 9.8  # Aceleracin de la gravedad (m/s)

#  Clculo de posicin, velocidad y aceleracin
y = v0 * t - (1/2) * g * t**2  # Posicin (m)
v = v0 - g * t  # Velocidad (m/s)
a = -g * np.ones_like(t)  # Aceleracin (m/s) (constante)

#  Grficos
fig, axs = plt.subplots(3, 1, figsize=(8, 10), sharex=True)
plt.suptitle("Grficos del Tiro Vertical", fontsize=16, fontweight='bold')

#  Grfico de posicin vs. tiempo
axs[0].plot(t, y, color='royalblue', linewidth=2)
axs[0].set_ylabel("Posicin (m)", fontsize=12)
axs[0].grid(True, linestyle='--', alpha=0.7)
axs[0].set_title("Posicin vs. Tiempo", fontsize=14)

#  Grfico de velocidad vs. tiempo
axs[1].plot(t, v, color='crimson', linewidth=2)
axs[1].set_ylabel("Velocidad (m/s)", fontsize=12)
axs[1].grid(True, linestyle='--', alpha=0.7)
axs[1].set_title("Velocidad vs. Tiempo", fontsize=14)

#  Grfico de aceleracin vs. tiempo
axs[2].plot(t, a, color='forestgreen', linewidth=2)
axs[2].set_xlabel("Tiempo (s)", fontsize=12)
axs[2].set_ylabel("Aceleracin (m/s)", fontsize=12)
axs[2].grid(True, linestyle='--', alpha=0.7)
axs[2].set_title("Aceleracin vs. Tiempo", fontsize=14)

#  Ajustes finales
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
