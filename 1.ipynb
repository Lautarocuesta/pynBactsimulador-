```python
import numpy as np
import matplotlib.pyplot as plt

# 🔹 Entrada de datos
h0 = float(input("Ingrese la altura inicial (m): "))
t_total = np.sqrt(2 * h0 / 9.8)  # Calculamos el tiempo total de caída
t = np.linspace(0, t_total, 100)  # Generamos puntos de tiempo en intervalos pequeños

# 🔹 Parámetros físicos
g = 9.8  # Aceleración de la gravedad (m/s²)

# 🔹 Cálculo de posición, velocidad y aceleración
y = h0 - (1/2) * g * t**2  # Posición (m)
v = -g * t  # Velocidad (m/s)
a = -g * np.ones_like(t)  # Aceleración (m/s²) (constante)

# 🔹 Gráficos
fig, axs = plt.subplots(3, 1, figsize=(8, 10), sharex=True)
plt.suptitle("Gráficos de Caída Libre", fontsize=16, fontweight='bold')

# 📉 Gráfico de posición vs. tiempo
axs[0].plot(t, y, color='royalblue', linewidth=2)
axs[0].invert_yaxis()  # Invertimos el eje para representar caída
axs[0].set_ylabel("Posición (m)", fontsize=12)
axs[0].grid(True, linestyle='--', alpha=0.7)
axs[0].set_title("Posición vs. Tiempo", fontsize=14)

# 📈 Gráfico de velocidad vs. tiempo
axs[1].plot(t, v, color='crimson', linewidth=2)
axs[1].set_ylabel("Velocidad (m/s)", fontsize=12)
axs[1].grid(True, linestyle='--', alpha=0.7)
axs[1].set_title("Velocidad vs. Tiempo", fontsize=14)

# ➡️ Gráfico de aceleración vs. tiempo
axs[2].plot(t, a, color='forestgreen', linewidth=2)
axs[2].set_xlabel("Tiempo (s)", fontsize=12)
axs[2].set_ylabel("Aceleración (m/s²)", fontsize=12)
axs[2].grid(True, linestyle='--', alpha=0.7)
axs[2].set_title("Aceleración vs. Tiempo", fontsize=14)

# 🔹 Ajustes finales
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 🔹 Entrada de datos
v0 = float(input("Ingrese la velocidad inicial (m/s): "))
t_total = (2 * v0) / 9.8  # Tiempo total de subida y bajada
t = np.linspace(0, t_total, 100)  # Generamos puntos de tiempo en intervalos pequeños

# 🔹 Parámetros físicos
g = 9.8  # Aceleración de la gravedad (m/s²)

# 🔹 Cálculo de posición, velocidad y aceleración
y = v0 * t - (1/2) * g * t**2  # Posición (m)
v = v0 - g * t  # Velocidad (m/s)
a = -g * np.ones_like(t)  # Aceleración (m/s²) (constante)

# 🔹 Gráficos
fig, axs = plt.subplots(3, 1, figsize=(8, 10), sharex=True)
plt.suptitle("Gráficos del Tiro Vertical", fontsize=16, fontweight='bold')

# 📉 Gráfico de posición vs. tiempo
axs[0].plot(t, y, color='royalblue', linewidth=2)
axs[0].set_ylabel("Posición (m)", fontsize=12)
axs[0].grid(True, linestyle='--', alpha=0.7)
axs[0].set_title("Posición vs. Tiempo", fontsize=14)

# 📈 Gráfico de velocidad vs. tiempo
axs[1].plot(t, v, color='crimson', linewidth=2)
axs[1].set_ylabel("Velocidad (m/s)", fontsize=12)
axs[1].grid(True, linestyle='--', alpha=0.7)
axs[1].set_title("Velocidad vs. Tiempo", fontsize=14)

# ➡️ Gráfico de aceleración vs. tiempo
axs[2].plot(t, a, color='forestgreen', linewidth=2)
axs[2].set_xlabel("Tiempo (s)", fontsize=12)
axs[2].set_ylabel("Aceleración (m/s²)", fontsize=12)
axs[2].grid(True, linestyle='--', alpha=0.7)
axs[2].set_title("Aceleración vs. Tiempo", fontsize=14)

# 🔹 Ajustes finales
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
