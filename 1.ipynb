```python
import numpy as np
import matplotlib.pyplot as plt

#  Entrada de datos
h0 = float(input("Ingrese la altura inicial (m): "))
t_total = float(input("Ingrese el tiempo total de simulacin (s): "))

#  Definir parmetros
g = 9.8  # Aceleracin de la gravedad (m/s)
t = np.linspace(0, t_total, 100)  # Generamos puntos de tiempo en intervalos pequeos

#  Clculo de posicin, velocidad y aceleracin
x = h0 - (1/2) * g * t**2  # Posicin (m)
v = -g * t  # Velocidad (m/s)
a = -g * np.ones_like(t)  # Aceleracin (m/s) (constante)

#  Grficos
plt.figure(figsize=(10, 6))

#  Grfico de posicin vs. tiempo
plt.subplot(3, 1, 1)
plt.plot(t, x, label='Posicin (m)', color='b')
plt.xlabel('Tiempo (s)')
plt.ylabel('Posicin (m)')
plt.title('Cada Libre - Posicin vs. Tiempo')
plt.grid()
plt.legend()

#  Grfico de velocidad vs. tiempo
plt.subplot(3, 1, 2)
plt.plot(t, v, label='Velocidad (m/s)', color='r')
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (m/s)')
plt.title('Cada Libre - Velocidad vs. Tiempo')
plt.grid()
plt.legend()

#  Grfico de aceleracin vs. tiempo
plt.subplot(3, 1, 3)
plt.plot(t, a, label='Aceleracin (m/s)', color='g')
plt.xlabel('Tiempo (s)')
plt.ylabel('Aceleracin (m/s)')
plt.title('Cada Libre - Aceleracin vs. Tiempo')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
