```python
import numpy as np
import matplotlib.pyplot as plt

# 🔹 Entrada de datos
h0 = float(input("Ingrese la altura inicial (m): "))
t_total = float(input("Ingrese el tiempo total de simulación (s): "))

# 🔹 Definir parámetros
g = 9.8  # Aceleración de la gravedad (m/s²)
t = np.linspace(0, t_total, 100)  # Generamos puntos de tiempo en intervalos pequeños

# 🔹 Cálculo de posición, velocidad y aceleración
x = h0 - (1/2) * g * t**2  # Posición (m)
v = -g * t  # Velocidad (m/s)
a = -g * np.ones_like(t)  # Aceleración (m/s²) (constante)

# 🔹 Gráficos
plt.figure(figsize=(10, 6))

# 📉 Gráfico de posición vs. tiempo
plt.subplot(3, 1, 1)
plt.plot(t, x, label='Posición (m)', color='b')
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (m)')
plt.title('Caída Libre - Posición vs. Tiempo')
plt.grid()
plt.legend()

# 📈 Gráfico de velocidad vs. tiempo
plt.subplot(3, 1, 2)
plt.plot(t, v, label='Velocidad (m/s)', color='r')
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (m/s)')
plt.title('Caída Libre - Velocidad vs. Tiempo')
plt.grid()
plt.legend()

# ➡️ Gráfico de aceleración vs. tiempo
plt.subplot(3, 1, 3)
plt.plot(t, a, label='Aceleración (m/s²)', color='g')
plt.xlabel('Tiempo (s)')
plt.ylabel('Aceleración (m/s²)')
plt.title('Caída Libre - Aceleración vs. Tiempo')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
